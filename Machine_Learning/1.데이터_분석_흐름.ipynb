# 데이터 수집

- 4차 산업의 원유
- 프로젝트가 성립되는 조건(단, 강화학습은 제외)
- 수집 방법
  - **level 1**
    - 데이터를 제공받음
    - 사내 데이터, 공공 데이터, 챌린지 데이터, 구매 데이터 등
  - level 2
    - open API 활용
    - 정규화된 데이터
    - 쿼터 제한
    - request
  - level 3
    - 웹 크롤링(웹 스크래핑)
    - BS4(html, xml 파싱)
    - request -> html 획득 -> BS4 -> DOM tree 생성 -> BS4를 통해 데이터 추출
    - 웹 크롤링시 법률 검토 필요
  - level 4
    - 로그인, ajax, coludflare 등 초기 페이지 방어 도구가 들어간 사이트
    - 쿠기를 활용한 사이트
    - 크롤링 : 자동화, 매크로, 봇

- 데이터 수집 자동화
  - level 2 이상, 웹에서 수집하는 경우
  - 수집 행위를 수행하는 파이썬 코드 작성
  - OS 레벨에서 자동화 생성
    - 윈도우 : 작업 스케줄러
    - 리눅스 : cron, crontab 명령
  - 데이터는 파일(csv), 데이터베이스|
- 웹 상에서 수집한 데이터는 상업적 이용시 법률적 검토 필요함

# 데이터 준비 / 전처리 / 정제

- 파이썬, 정규식, 판다스(넘파이) <--> R
- 분석이나 학습에 사용 가능하도록 데이터를 전처리하는 과정
- 양질의 데이터를 생성하는 과정
- 데이터의 품질을 향상시킴
- 머신러닝과 딥러닝에서 요구하는 데이터 형태
  - 수치화를 목표로 전처리
  - 정규화 부분 고려
  - 딥러닝을 기본적으로 행렬 연산을 근간으로 함
- 결측치, 이상치 처리
- 독립변수(feature, 특성), **종속변수(label, class, 정답)**, 파생변수(계산 후 생성)

# 데이터 분석

- R <--> 판다스(넘파이 기반), 기타 모듈
- 파이썬으로 진행하는 이유는?
  - 산출물에 대한 스펙트럼이 넓음
  - 딥러닝, 웹 등 다양한 형태로 확장 가능
- 자격증
  - ADSP < 빅데이터 분석기사 < ADP
  - 자격증의 시험 범위는 여기까지임
- 기법
  - 설명 분석
  - 탐색 분석
  - 추론 분석
  - 인과 분석
  - 규범 분선
  - 예측 분석 : 머신러닝에 해당
- 머신러닝과 딥러닝 관점에서는
  - 데이터를 다음 단계로 진행해도 합당한지를 검증하는 단계(머신러닝)
  - 딥러닝에서는 거의 생략됨

# 모델 구축

- 데이터 모델 / 예측 모델 / 통계 모델
- 머신러닝 / 딥러닝 모델 학습
- 머신러닝(기계학습)
  - 지도 학습(문제 해결 초점, 종속변수(정답) 있음)
    - 분류
      - 이 이미지는 고양이이다
      - 이것은 이거다 처럼 정확한 답
      - 대부분의 딥러닝의 주제와 부합
    - 회귀
      - 이 아파트는 1년 후 30억원 전후로 상승할 것이다(집값 예측)
      - 어느정도 범위가 존재하는 답을 제시
  - 비지도 학습(전처리와 데이터의 특징을 캐치 초점, 종속변수(정답) 없음)
    - 군집
      - 추천 시스템의 데이터 구축
        - 사용자 유형 정의
        - 쇼핑몰 기준 비슷한 유형의 소비자들을 정의(다양한 데이터를 근거로)
    - 차원축소
      - 독립변수가 많아서 몇 개로 압축해서 새로운 파생변수를 생성
      - 상관성 등 고려
  - 준지도 학습
    - 지도 + 비지도
    - 데이터(중 종속변수)가 일부는 있고 일부는 없음
  - 강화 학습
    - 특정 환경에서 보상을 잘 받는 행동을 강화시키도록 에이전트를 학습(훈련)하는 학습법
    - 환경(ex 게임의 룰), 에이전트(ex 플레이어), 보상, 정책, 정책갱신, 행동, 패널티 등 이런 요소를 정의하고 학습시키는 방식
    - 강화학습 + 시뮬레이션(유니티 등) 분야 확장(ML - Agent)
    - 알파고, 알파고제로, 알파제로, 알파스타(스타크래프트)), 알파폴드(단백질유전자))

- 딥러닝
  - 장비 : GUP
  - 엔진
    - 텐서플로우 2.x 기반 + 케라스
      - 텐서플로우 : 내부까지 건드림
      - 케라스 : 인터페이스 레벨에서 작성
    - 파이토치(국책연구기관)
  - 분야
    - 이미지, 영상, 객체 탐지(싱글, 멀티)
      - CNN, CNN 파생 모델(ResNet, VGC), 써드타입(YoLo, MediaPipe)
    - 자연어
      - 텍스트마이닝, 문서 요약, 문서 생성, 기계번역, 챗봇
      - RNN, RNN - LSTM, RNN - GRU, 자연어 처리 모듈 등
    - 시계열
      - 주식, 금융, 스마트 팩토리 등 분야에서 나오는 데이터
      - 데이터에 시간, 순서가 있음
      - RNN, RNN -LSTM, RNN - GRU
    - 생성모델
      - 딥페이크(이미지 생성, 채색, 영상 생성, 음악 생성)
      - GAN
      - 학습시간이 오래 걸럼
    - 강화학습 분야
